```python
import fitz  # PyMuPDF
import pandas as pd
import re

# Step 1: Load the PDF
doc = fitz.open("opaywalletstatement.pdf")  # Replace with your actual PDF path

# Step 2: Extract text from each page
text = ""
for page in doc:
    text += page.get_text()

# Step 3: Use regex to extract transaction rows
pattern = re.compile(
    r"(\d{4} \w{3} \d{2} \d{2}:\d{2}:\d{2})\s+(\d{2} \w{3} \d{4})\s+(.*?)\s+([+-]?\d{1,3}(?:,\d{3})*(?:\.\d{2})?)\s+(\d{1,3}(?:,\d{3})*(?:\.\d{2})?)\s+E-Channel\s+(\d+)"
)

transactions = pattern.findall(text)

# Step 4: Convert to DataFrame
df = pd.DataFrame(transactions, columns=[
    "Transaction Time", "Value Date", "Description", "Amount", "Balance", "Transaction Reference"
])

# Step 5: Convert numbers to float
df["Amount"] = df["Amount"].str.replace(",", "").astype(float)
df["Balance"] = df["Balance"].str.replace(",", "").astype(float)

# Step 6: Save to Excel
df.to_excel("opay_statement.xlsx", index=False)
```
